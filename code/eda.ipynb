```python
#Calculate enrollment changes in the 5 boroughs from the start year of 2015-16 to the end year 2019-20

def calculate_boro_enrollment_change(boro_enrollment):
    changes = {}
    
    boroughs = boro_enrollment["Borough"].unique()
    for boro in boroughs:
        boro_df = boro_enrollment[boro_enrollment['Borough'] == boro]
        enrollment_2015 = boro_df[boro_df['Year'] == '2015-16']['Total Enrollment'].sum()
        enrollment_2020 = boro_df[boro_df['Year'] == '2019-20']['Total Enrollment'].sum()
        changes[boro] = (enrollment_2020 - enrollment_2015)
    
    return changes

changes = calculate_boro_enrollment_change(boro_enrollment)
for borough, change in changes.items():
    print(f"{borough}: Change in total enrollment between 2015 and 2020 = {change} students\n")
```

___
> **Insights on Total Enrollment**

- Citywide K-12 enrollment was highest during the 2016-2017 school year
- The lowest total K-12 enrollment academic year was 2018-2019 with 1,126,501 students
- Citywide enrollment during the 2019-2020 school year - 1,131,868 students - increased from the previous year but did not meet the 2015-2016 headcount of 1,133,963 students.

**[The Economic Need Index Metric](https://data.cccnewyork.org/data/bar/1371/student-economic-need-index#1371/a/1/1622/127)**

**The schoolâ€™s Economic Need Index is the average of its studentsâ€™ Economic Need Values. The Economic Need Index (ENI) estimates the percentage of students facing economic hardship. The metric is calculated as follows:**

**The studentâ€™s Economic Need Value is 1.0 if:**
- The student is eligible for public assistance from the NYC Human Resources Administration (HRA);
- The student lived in temporary housing in the past four years; or
- The student is in high school, has a home language other than English, and entered the NYC DOE for the first time within the last four years.

 **Otherwise, the studentâ€™s Economic Need Value is based on the percentage of families (with school-age children) in the studentâ€™s census tract whose income is below the poverty level, as estimated by the American Community Survey 5-Year estimate (2016 ACS estimates were used in calculations for 2018-19 ENI). The studentâ€™s Economic Need Value equals this percentage divided by 100.**

```python
#Inspect total Economic Need Index throughout 2015-2020

economic_need_index_df = df[["Year","Economic Need Index"]]
economic_need_index_df
```

```python
#Visualize the ENI df
plt.figure(figsize=(20, 18))
economic_need_index_df.plot.line(x="Year", y="Economic Need Index", linewidth=5.0)

#Remove scientific notation from y axis
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False) 
plt.grid()
plt.title("\nEconomic Need Index (EIN) of NYC Public Schools from 2015-2020\n", fontsize=16, weight='bold', style='italic')
plt.xlabel("\nYear\n", fontsize=14, weight='bold')
plt.ylabel("\nEconomic Need Index (EIN)\n", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

___
> **Insights on Economic Need Index**

- From 2015-2017, the overall Economic Need Index remains somewhat consistent at 0.621 and 0.61
- The academic year range of 2017 - 2020 shows an increase of Economic Need Index increase from 0.61 to a steady 0.71.
