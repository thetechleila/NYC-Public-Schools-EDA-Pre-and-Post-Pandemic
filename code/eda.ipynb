```python
#Examine total student enrollment for the 5 school year ranges
total_enrollment_df = df[["Year","Total Enrollment"]]
total_enrollment_df 
```

```python
#Get enrollment across the 5 Boroughs from the 2015-16 to 2019-20 school years

boro_enrollment  = boro_dem_df[["Year", "Borough", "Total Enrollment"]]
boro_enrollment
```

```python
#Visualize the total student enrollment df

plt.figure(figsize=(16, 10))
sns.lineplot(data=total_enrollment_df, x="Year", y="Total Enrollment", marker="o", linewidth = 9.0)

#Remove scientific notation from y axis
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False) 
plt.grid(color="black", linewidth=0.5)
plt.title("\nTotal Citywide Enrollment in NYC Public Schools 2015 - 2020\n", fontsize=16, 
         weight='bold', style='italic')
plt.xlabel("\nYear\n", fontsize=14, weight='bold')
plt.ylabel("\nTotal Enrollment\n", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

```python
#Visualize student enrollment by borough from 2015-2020

plt.figure(figsize=(16, 10))
sns.lineplot(data=boro_enrollment, x="Year", y="Total Enrollment", hue="Borough", marker="o", linewidth = 9.0)
plt.title("\nNYC Public School Enrollment by Borough 2015-2020\n", 
          weight="bold", style="italic")

#Plot labels and customizations
plt.xlabel("\nAcademic Year\n", weight='bold' )
plt.ylabel("Total Enrollment\n", weight='bold')
plt.grid(color="black", linewidth=0.5)
plt.legend(title="Grade Level")

plt.tight_layout()
plt.show()
```

```python
#Calculate enrollment changes in the 5 boroughs from the start year of 2015-16 to the end year 2019-20

def calculate_boro_enrollment_change(boro_enrollment):
    changes = {}
    
    boroughs = boro_enrollment["Borough"].unique()
    for boro in boroughs:
        boro_df = boro_enrollment[boro_enrollment['Borough'] == boro]
        enrollment_2015 = boro_df[boro_df['Year'] == '2015-16']['Total Enrollment'].sum()
        enrollment_2020 = boro_df[boro_df['Year'] == '2019-20']['Total Enrollment'].sum()
        changes[boro] = (enrollment_2020 - enrollment_2015)
    
    return changes

changes = calculate_boro_enrollment_change(boro_enrollment)
for borough, change in changes.items():
    print(f"{borough}: Change in total enrollment between 2015 and 2020 = {change} students\n")
```

___
> **Insights on Total Enrollment**

- Citywide K-12 enrollment was highest during the 2016-2017 school year
- The lowest total K-12 enrollment academic year was 2018-2019 with 1,126,501 students
- Citywide enrollment during the 2019-2020 school year - 1,131,868 students - increased from the previous year but did not meet the 2015-2016 headcount of 1,133,963 students. 

___

> **Insights on *Borough* Enrollment**

- K-12 Student Enrollment rates throughout Staten Island, Brooklyn, and Queens remained steady from 2015-2020 

- The most notable changes in K - 12 enrollment occurred in The Bronx and Manhattan
    - The Bronx lost 6538 students. Overall enrollment in the Borough decreased every year from 2015 to 2020.

**[The Economic Need Index Metric](https://data.cccnewyork.org/data/bar/1371/student-economic-need-index#1371/a/1/1622/127)**

**The school’s Economic Need Index is the average of its students’ Economic Need Values. The Economic Need Index (ENI) estimates the percentage of students facing economic hardship. The metric is calculated as follows:**

**The student’s Economic Need Value is 1.0 if:**
- The student is eligible for public assistance from the NYC Human Resources Administration (HRA);
- The student lived in temporary housing in the past four years; or
- The student is in high school, has a home language other than English, and entered the NYC DOE for the first time within the last four years.

 **Otherwise, the student’s Economic Need Value is based on the percentage of families (with school-age children) in the student’s census tract whose income is below the poverty level, as estimated by the American Community Survey 5-Year estimate (2016 ACS estimates were used in calculations for 2018-19 ENI). The student’s Economic Need Value equals this percentage divided by 100.**

```python
#Inspect total Economic Need Index throughout 2015-2020

economic_need_index_df = df[["Year","Economic Need Index"]]
economic_need_index_df
```

```python
#Visualize the ENI df
plt.figure(figsize=(16, 10))
sns.lineplot(data=economic_need_index_df, x="Year", y="Economic Need Index", 
             marker="o", linewidth = 9.0)
 

plt.grid(color="black", linewidth=0.5)
plt.title("\nEconomic Need Index (EIN) of NYC Public Schools from 2015-2020\n", 
          fontsize=16, weight='bold', style='italic')
plt.xlabel("\nSchool Year\n", fontsize=14, weight='bold')
plt.ylabel("\nEconomic Need Index (EIN)\n", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```
