**[The Economic Need Index Metric](https://data.cccnewyork.org/data/bar/1371/student-economic-need-index#1371/a/1/1622/127)**

**The school’s Economic Need Index is the average of its students’ Economic Need Values. The Economic Need Index (ENI) estimates the percentage of students facing economic hardship. The metric is calculated as follows:**

**The student’s Economic Need Value is 1.0 if:**
- The student is eligible for public assistance from the NYC Human Resources Administration (HRA);
- The student lived in temporary housing in the past four years; or
- The student is in high school, has a home language other than English, and entered the NYC DOE for the first time within the last four years.

 **Otherwise, the student’s Economic Need Value is based on the percentage of families (with school-age children) in the student’s census tract whose income is below the poverty level, as estimated by the American Community Survey 5-Year estimate (2016 ACS estimates were used in calculations for 2018-19 ENI). The student’s Economic Need Value equals this percentage divided by 100.**

```python
#Inspect total Economic Need Index throughout 2015-2020

economic_need_index_df = df[["Year","Economic Need Index"]]
economic_need_index_df
```

```python
#Visualize the ENI df
plt.figure(figsize=(20, 15))
economic_need_index_df.plot.line(x="Year", y="Economic Need Index")

#Remove scientific notation from y axis
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False) 
plt.grid()
plt.title("\nEconomic Need Index (EIN) of NYC Public Schools from 2015-2020\n", fontsize=16, weight='bold', style='italic')
plt.xlabel("\nYear\n", fontsize=14, weight='bold')
plt.ylabel("\nEconomic Need Index (EIN)\n", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

___
> **Insights on Economic Need Index**

- From 2015-2017, the overall Economic Need Index remains somewhat consistent at 0.621 and 0.61
- The academic year range of 2017 - 2020 shows an increase of Economic Need Index increase from 0.61 to a steady 0.71.

```python
#Get overall male and female NYC public school student enrollment for the time range
gender_df = df[["Year", "Total Enrollment", "# Female", "# Male"]]
gender_df

```

```python
# Set variables & data for grouped gender bar chart
years = gender_df['Year']
female_count = gender_df['# Female']
male_count = gender_df['# Male']

# Define bar position and width
x = np.arange(len(years))  
width = 0.35  

# Visualize grouped bar chart of female vs male K-12 NYC Public School students from 2015-2020
fig, ax = plt.subplots(figsize=(16, 10))
bars1 = ax.bar(x - width/2, female_count, width, label='Female', color='darkorange')
bars2 = ax.bar(x + width/2, male_count, width, label='Male', color='steelblue')


ax.set_ylabel('Number of Students\n', weight='bold')
ax.set_xlabel('\n Academic Year\n', weight='bold')
ax.set_title('Total NYC K-12 Female & Male Public School Student Enrollment 2015-2020\n', weight='bold', style='italic')

ax.set_axisbelow(True) 
ax.grid(True)
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.legend()

plt.tight_layout()
plt.show()
```

___
> **Insights on the total enrollment of female and male students in NYC from the 2015-2020 Academic Years**

- Overall, there are usually slightly more total male K-12 students than total female K-12 students enrolled in the NYC Public School system from 2015 - 2020.
- The close ratio of total female to total male K-12 student enrollment in the NYC Public School remained relatively consistent throughout each 5 of the academic school years examined.

**High School (secondary school) in the New York City Public School System contains Grades 9 - 12 and is traditionally completed in 4 years.**
___

```python
# Get total high school enrollment rates from 2015-2020
hs_enrollment_df = df[["Year", "Grade 9", "Grade 10", "Grade 11", "Grade 12"]]
hs_enrollment_df
```

```python
# Melt hs enrollment df into long format 
df_melt = hs_enrollment_df.melt(id_vars="Year", 
                                   var_name="Grade", 
                                   value_name="Enrollment")


#Visualize Total NYC High School Student Enrollment 2015-2020

plt.figure(figsize=(16, 8))
sns.lineplot(data=df_melt, x="Year", y="Enrollment", hue="Grade", marker="o")
plt.title("\nNYC Public High School Total Enrollment Rates 2015-2020\n", 
          weight="bold", style="italic")

#Plot labels and customizations
plt.xlabel("\nAcademic Year\n", weight='bold' )
plt.ylabel("Total Enrollment\n", weight='bold')
plt.grid(True)
plt.legend(title="Grade Level")

plt.tight_layout()
plt.show()
```

___
> **Insights on the total enrollment rates for high school students in NYC from 2015-2020**

- 9th grade enrollment decreased from 91723	students during the 2015-2016 school year to 88773 students during the 2018-2019 year with only a sligth increase in the following year.
- 11th and 12th grade student enrollment is consistently approximately 10,000 - 23,000 students fewer than 9th and 10th grade enrollment from 2015-2020.
- 10th grade student enrollment has steadily decreased every school year from 2015 to 2020
    - There are 90854 students enrolled in 10th grade in 2015-16 
    - There are 86802 students enrolled in 10th grade in 2019-20
